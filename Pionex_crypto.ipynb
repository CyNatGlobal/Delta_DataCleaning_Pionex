```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import re
from datetime import datetime

# Display all columns
pd.set_option('display.max_columns', None)

# Step 2: Load Data and Standardize Column Names
from google.colab import files
uploaded = files.upload()  # Upload the file
data = pd.read_csv('/content/241k-Singapore-pionex.com-Crypto-Trading-Bots-UsersDB-csv-2023.csv')  # Load the data

# Replace spaces in headers with underscores
data.columns = data.columns.str.replace(' ', '_')
print("Updated column names:", data.columns)

# Step 3: Set Up Junk DataFrame for Invalid Entries
junk_data = pd.DataFrame(columns=data.columns)

# Step 4: Validate Emails
email_pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

# Identify and move invalid email rows to junk_data
invalid_emails = data[~data['Email'].str.match(email_pattern, na=False)]
junk_data = pd.concat([junk_data, invalid_emails])
data = data[data['Email'].str.match(email_pattern, na=True)]

# Step 5: Validate Registration Dates
def validate_date(date):
    try:
        return pd.to_datetime(date)
    except:
        return pd.NaT

data['RegistrationDate'] = data['RegistrationDate'].apply(validate_date)

# Separate invalid or missing dates to junk_data
invalid_dates = data[data['RegistrationDate'].isnull()]
junk_data = pd.concat([junk_data, invalid_dates])
data.dropna(subset=['RegistrationDate'], inplace=True)

# Step 6: Validate Phone Numbers


# Separate invalid phone numbers to junk_data


# Step 7: Remove Duplicates
duplicates = data[data.duplicated()]
junk_data = pd.concat([junk_data, duplicates])
data.drop_duplicates(inplace=True)

# Step 8: Handle Remaining Missing Values
remaining_missing = data[data.isnull().any(axis=1)]
junk_data = pd.concat([junk_data, remaining_missing])
data.dropna(inplace=True)

# Step 9: Save Cleaned Data and Junk File
data.to_csv('cleaned_data.csv', index=False)
junk_data.to_csv('junk_data.csv', index=False)

# Download the files to your local system
from google.colab import files
files.download('cleaned_data.csv')
files.download('junk_data.csv')
```
